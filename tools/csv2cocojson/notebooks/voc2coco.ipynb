```python
import json
```

```python
def bbox2mask(bbox):
    return [bbox[0], bbox[1], bbox[0]+bbox[2], bbox[1], bbox[0]+bbox[2], bbox[1]+bbox[3], bbox[0], bbox[1]+bbox[3]]
```

```python
with open('instances_train2017.json') as f:
    df = json.load(f)
    
for i in range(len(df['annotations'])):
    bbox = df['annotations'][i]['bbox']
    df['annotations'][i]['segmentation'] = [bbox2mask(bbox)]
    #print(df['annotations'][i]['segmentation'])
    #break
    
with open('instances_train2017.json', 'w') as f:
    json.dump(df, f, indent=4)
```

```python
with open('instances_val2017.json') as f:
    df = json.load(f)
    
for i in range(len(df['annotations'])):
    bbox = df['annotations'][i]['bbox']
    df['annotations'][i]['segmentation'] = [bbox2mask(bbox)]
    #print(df['annotations'][i]['segmentation'])
    #break
    
with open('instances_val2017.json', 'w') as f:
    json.dump(df, f, indent=4)
```
