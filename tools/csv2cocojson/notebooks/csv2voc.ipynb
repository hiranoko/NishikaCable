```python
# modules
# https://github.com/AndrewCarterUK/pascal-voc-writer

import os
import sys
import glob
import tqdm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import cv2

sys.path.append("/home/hirano/work/Competition/Nishika_cable/csv2cocojson/pascal-voc-writer/")
import pascal_voc_writer

%matplotlib inline
```

```python
def as_list_bbox(bbox: str):
    return np.array(bbox.replace("[", "").replace("]", "").split(","), dtype=np.uint16)

def viz_annotation(imgname, bbox):
    imgname = row['filename']
    bbox = as_list_bbox(row['Xmin_Ymin_Xmax_Ymax'])
    img = cv2.imread("../../input/train/{}".format(imgname))
    img = cv2.rectangle(img, bbox[0:2], bbox[2:4], (0,255,0), 12)
    plt.imshow(img[:, :, ::-1])
    plt.show()
    
def create_annotation(img_name, bbox, class_name):
    img_name = row['filename']
    bbox = as_list_bbox(row['Xmin_Ymin_Xmax_Ymax'])
    class_name = row['class']
    img = cv2.imread("../../input/train/{}".format(img_name))
    height, width = img.shape[:2]
    
    writer = pascal_voc_writer.Writer("../../input/train/{}".format(img_name), width, height)
    writer.addObject(name=class_name,xmin=bbox[0],ymin=bbox[1],xmax=bbox[2],ymax=bbox[3]) 
    writer.save('../output/{}.xml'.format(img_name.split('.')[0]))
```
